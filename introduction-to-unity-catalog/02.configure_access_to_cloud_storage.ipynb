## Configure Access to Cloud Storage  via Unity Catalog

**Access Cloud Storage(Storage Container)**

```python
%fs ls 'abfss://dev@databrickskrogerextdl.dfs.core.windows.net/'
```

**Create External Location For Storage Container**

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrickskrogerextdl_dev
URL 'abfss://dev@databrickskrogerextdl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `kroger-dev-sc`)
COMMENT 'External Location For Dev Container';
```
